# Convert Celsius to Fahrenheit

In this project we are going to do a very simple neuronal network. The goal is to train a model to convert from Celsius degrees to Fahrenheit degrees. 

First let's import some libraries and create our data training. It's formed by two arrays, one with the Celsius value and other with the equivalent in Fahrenheit.

```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

Create the dense layer with 1 neuron and one imput layer with 1 neuron too.

```python
layer = tf.keras.layers.Dense(units=1, input_shape=[1])
model = tf.keras.Sequential([layer])
```

Train the model with Adam algorithm and Mean Squared Error.

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Training...")
history = model.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Finished training")
```

Let's see the loss in each epoch

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

As we can see the full training is done by the epoch 500. Let's do our first prediction.

```python
print("Predicting...")
value = 100.0
prediction = model.predict([value])
print("Prediction: {} Celsius is {} Fahrenheit".format(value, prediction))
```

The exact convertion from 100 Celsius degrees is 212 Fahrenheit degrees. The structure of the network is the following:

```python
layer_weights = layer.get_weights()
print("Layer weights: {}".format(layer_weights))
```

The formula to convert from Celsius to Fahrenheit is:
fahrenheit = celsius * 1.8 + 32

Our models uses 1.7981397 instead of 1.8 and 31.930012 instead of 32, so we conclude that the models is well trained.
